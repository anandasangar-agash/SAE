<h5>Exercice</h5>
<p>Ecrire des fonctions permettrant de calculer la <code>moyenne</code>, la <code>covariance</code>, la <code>variance</code> et <code>l'écart-type</code> d'une liste de donnée numérique <code>x</code> (et <code>y</code> pour la covariance.<br>
La variance et l'écart type sont déduits de la covariance de x et y.</p>

```python
def moyenne(x) : 
    try : n=len(x)
    except : return None
    if(n==0) : return None
    
    res = 0
    i = 0
    som = 0
    while i < len(x):
        som = som + x[i]
        i =i+ 1
    res = som / len(x)    
    return res
```

```python
def covariance(x, y) :
    try : n=len(x)
    except : return None
    if(n==0 or n!=len(y)) : return None
    i = 0
    a = np.zeros_like(x)
    while i < len(x):
        a[i] = x[i]*y[i]
        i = i+1
    b = moyenne(x)*moyenne(y)
    cov = moyenne(a) - b
    return cov
```

```python
def variance(x) : 
    res = covariance(x,x)
    return res
```

```python
def ecartype(x) : 
    res= sqrt(variance(x))
    return res
```

```python
#Test
print("La moyenne du JEU est", round(moyenne(JEU),2))#~2.37
print("L'écart-type du JEU est", round(ecartype(JEU),2))#~1.46
```

<h3>Code</h3>
Compléter le programme pour écrire la fonction <code>r(x, y)</code> qui calcule le coefficient de corrélation linéaire de <b>Pearson </b>
</p>

```python
def r(x,y):
    try:
        n=len(x)
    except:
        return None
    if(n==0 or n!=len(y)):
        return None
    if ecartype(x)==0 or ecartype(y)==0:
        return None
    else: 
        a = covariance(x,y)
        b = ecartype(x)*ecartype(y)
        return a / b
```

<h3>Code</h3>
Ecrivez les fonctions suivantes
<ul>
    <li><code>achapeau(x, y)</code> qui calcule le coefficient directeur de la droite de régression linéaire</li>
    <li><code>bchapeau(x, y)</code> qui calcule l'ordonnée à l'origine de la droite de régression linéaire</li>
    <li><code>ychapeau(x, y)</code> qui renvoie la liste des estimations du modèle ($\hat{y}_i=\hat{a}x_i+\hat{b}$) </li>
    <li><code>residu(x, y)</code> qui renvoie la liste des résidu du modèle ($\hat{\varepsilon}_i=\hat{y}_i-y_i$)</li>
</ul>
</p>

```python
def achapeau(x, y) : return covariance(x,y)/variance(x)
```

```python
def bchapeau(x, y) : return moyenne(y) - (achapeau(x,y)*moyenne(x))
```

```python
def ychapeau(x, y) :
    n=len(x)
    if(len(y)!=n or n==0) : return None
    res=[]
    a=achapeau(x,y)
    b=bchapeau(x,y)
    i = 0
    while i < len(x):
        res.append(a*x[i]+b)
        i += 1
    return res
ychapeau(Revenu,Consommation)
```

```python
def residu(x, y) : 
    n=len(x)
    if(len(y)!=n or n==0) : return None
    res=[]
    i = 0
    while i < len(x):
        res.append(x[i]-y[i])
        i += 1
    return res
a = ychapeau(Revenu,Consommation)
residu(a,Consommation)
```

```python
#Extractions des notes
nbNote = 6
NOTES=[]
for i in range(nbNote) : 
    NOTES.append([])
for etudiant in JEU : 
    i=0
    while(i<nbNote) : 
        NOTES[i].append(JEU[etudiant][i])
        i+=1
        
#Analyse
#Un modèle sera dit "de qualité" si R^2>0.6
Qual=60/100
for i in range(len(NOTES)):
    for j in range(i+1,len(NOTES)):
        print("--------: ANALYSE DE CORRELATION LINEAIRE ENTRE LES NOTES",i+1 ,"ET", j+1, " :--------")
        
#Analyse en dimesnion 1
        print("\n\tMoyenne notes", i+1, ":", moyenne(NOTES[i]))
        print("\tEcart-type des notes", i+1, ":", ecartype(NOTES[i]))
        print("\n\tMoyenne notes", j+1, ":", moyenne(NOTES[j]))
        print("\tEcart-type des notes", j+1, ":", ecartype(NOTES[j]))
        
#Analyse en dimension 2
        print("\n\tCovariance entre les notes",i+1 ,"et", j+1, ":", covariance(NOTES[i],NOTES[j]))
        
#Nuage de points
        scatter(NOTES[i],NOTES[j])
        grid(True)
        xlabel("Notes "+str(i+1))
        ylabel("Notes "+str(j+1))
        xlim(0, 20)
        ylim(0, 20)
        R = r(NOTES[i],NOTES[j])
        print("\n\n\tQualité du modèle : r^2 =", R**2)
        
        if(R**2>Qual) :
#Droite de régression linéaire
            plot(R,'r', label="Estimation") 
            legend()
            
#coef de la droite
            print("\tachapeau =", achapeau(NOTES[i],NOTES[j]))
            print("\tbchapeau =", bchapeau(NOTES[i],NOTES[j]))
        else : 
            print("\tLa qualité du modèle n'est pas bonne")
            show()
```
